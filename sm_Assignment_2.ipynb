## Data Visualization

#### Univariate Analysis

```python
plt.rcParams['figure.figsize'] = 12,5
sns.histplot(data['living area'],kde=True)
plt.title('Living Area')
plt.show()
```

```python
plt.plot(data['number of bathrooms'])
plt.title('Number of Bathrooms')
plt.show()
```

```python
pie_data=data['Number of schools nearby'].value_counts()
plt.pie(pie_data,autopct='%1.2f%%',labels=['3 Schools','2 Schools','1 School'])
plt.legend(loc='right')
plt.title('Number of schools nearby')
plt.show()
```

#### Bi-variate Analysis

```python
plt.bar(data['number of bedrooms'],data['Price'])
plt.xlabel('Number of bedrooms')
plt.ylabel('Price')
plt.title('Comparsion of bedrooms with price')
plt.show()
```

```python
plt.scatter(data['number of floors'],data['lot area'])
plt.xlabel('Number of floors')
plt.ylabel('lot area')
plt.title('Comparsion of floors with lot area')
plt.show()
```

```python
sns.jointplot(y=data['number of floors'],x=data['number of bedrooms'])
plt.ylabel('Number of bedrooms')
plt.ylabel('Number of floors')
plt.show()
```

## Multivariate Analysis

```python
#Taking short data for better representation
short_data=data[['number of bedrooms','number of bathrooms','living area','number of floors','Price']]
```

```python
sns.pairplot(short_data)
plt.show()
```

```python
plt.rcParams['figure.figsize']=15,8
correlation=short_data.corr()
sns.heatmap(correlation,annot=True)
plt.show()
```
